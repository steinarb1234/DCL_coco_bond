```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

total_debt =  274_280_000_000 
market_cap = 48_813_335_162
# Plot for different L_min, from 0 to 1 with 0.01 step
L_min = np.arange(0.85, .95, 0.01)
# L_min = 0.85

L_k = total_debt / (total_debt + market_cap)
L_min_distance = L_min - L_k # distance to L_min

# Additional loan size to adjust L up to L_min with top-up loan
delta_D = -total_debt + L_min * market_cap / (1 - L_min)

# print(f"Q_k: {Q_k:,.2f}")

# Additional loan size to adjust L up to L_min with top-up loan and share buybacks with the loan
delta_D_share_buybacks =  -total_debt + L_min * (total_debt + market_cap)

# print(f"Q_k with share buybacks: {Q_k_share_buybacks:,.2f}")
# print(f"Share buyback value (total value of equity drawdown): {Q_k_share_buybacks:,.2f}")

```

```python
# Plotting

plt.figure(figsize=(10, 6))
plt.plot(L_min, delta_D, label='Q_k', color='blue')

plt.title(f'Top up loan size to adjust L = {L_k:.4f} up to L_min')
plt.xlabel('L_min')
plt.ylabel('Q_k')
plt.grid()
# plt.axhline(0, color='black', lw=0.5, ls='--')
# plt.axvline(0, color='black', lw=0.5, ls='--')
# plt.axvline(1, color='black', lw=0.5, ls='--')
# Show values with 2 decimal points
plt.xticks(L_min, rotation=45)
# Show values on the line
# for i, j in zip(L_min, delta_D):
#     plt.text(i-0.005, j, f'{j:,.0f}', fontsize=8, ha='center', va='bottom', rotation=0)

plt.legend()
plt.tight_layout()
plt.savefig('images/Top up loan size to adjust L up to L_min.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Plotting

plt.figure(figsize=(10, 6))
plt.plot(L_min, delta_D, label='Q_k', color='blue')
plt.plot(L_min, delta_D_share_buybacks, label='Q_k with share buybacks', color='orange')

plt.title(f'Top up loan size to adjust L = {L_k:.4f} up to L_min')
plt.xlabel('L_min')
plt.ylabel('Q_k')
plt.grid()
# plt.axhline(0, color='black', lw=0.5, ls='--')
# plt.axvline(0, color='black', lw=0.5, ls='--')
# plt.axvline(1, color='black', lw=0.5, ls='--')
# Show values with 2 decimal points
plt.xticks(L_min, rotation=45)
# Show values on the line
# for i, j in zip(L_min, delta_D):
#     plt.text(i, j, f'{j:,.0f}', fontsize=8, ha='center', va='bottom', rotation=0)
# for i, j in zip(L_min, delta_D_share_buybacks):
#     plt.text(i, j, f'{j:,.0f}', fontsize=8, ha='center', va='bottom', rotation=0)

plt.legend()
plt.tight_layout()
plt.savefig('images/Top up loan size to adjust L up to L_min share buybacks.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

output_df = pd.DataFrame({
    'L_min': L_min,
    'distance': L_min_distance,
    'Q_k': delta_D,
    'Q_k_share_buybacks': delta_D_share_buybacks
})

print(output_df)
```
